### Dokumentera:

#### Preliminär beskrivning på hur uppgiften ska utföras.

	Upplägg med teknik och teknologier som används ska i projektet.
	Testdata ska simuleras och bestämmas i utvecklingsverktyget Node-Red som fungerar som en publiserare av data med hjälp av MQTT protokollet. Datainnehållet ska vara av IoT typ. Antal parametrar kan vara flera och av olika typer. För att skapa realitet i projektet.
	Datan mottages av en broker(server som fördelar datan till prenumeranterna) kan vara lokal via Eclipse Mosquitto eller annan passande programvara. En public broker kan lätt användas som alternativ, då ip adressen enkelt kan ändras i applikationen.
	Data ska också sparas i en time scale databas typ influxdb. Den är ett vanligt alternativ inom IoT applikationer och är väl integrerbar med Grafana som används som visualiseringsverktyg.
	I Grafana ska också Alert managern användas för att sätta de tänkta gränsvärdena som ska förutsättas.
	En machine learning applikation med en regressionsmodel ska skrivas i Python. Denna betraktas även som prenumerant av MQTT data som erhålls från brokern. Resultatet ska innehålla förutsägelserna.
	Virtualiseringsteknik ska användas, Virtualbox och Docker.
